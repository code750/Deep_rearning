# 보스턴 집값 예측

*   github url : https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv



```python
# 데이터 준비
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
보스턴 = pd.read_csv(파일경로)
print(보스턴.columns)
보스턴.head()
```

```python
# 독립변수, 종속변수 분리
독립 = 보스턴[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
종속 = 보스턴[['medv']]
print(독립.shape, 종속.shape)
```

```python
# 모델 구조 만들기
x = tf.keras.layers.Input(shape=13)
h = tf.keras.layers.Dense(10, activation='swish')(x)
y = tf.keras.layers.Dense(1)(h)
model = tf.keras.models.Model(x, y)
model.compile(loss='mse')
```

```python
model.summary()
```

```python
# 모델 학습시키기
model.fit(독립, 종속, epochs=500)
```

```python
# 모델 이용하기
print(model.predict(독립[0:5]))
print(종속[:5])
```

```python
종속[5:10]
```

```python
# 모델의 수식 확인
model.get_weights()
```

# 아이리스 품종 분류

*   github url : https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv



```python
# 라이브러리 가져오기
import tensorflow as tf
import pandas as pd
```

```python
# 데이터 준비
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
아이리스 = pd.read_csv(파일경로)
아이리스.head()
```

```python
# 인핫 인코딩
인코딩 = pd.get_dummies(아이리스)
인코딩.head()
```

```python
print(인코딩.columns)
```

```python
# 독립변수, 종속변수 만들기
독립 = 인코딩[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속 = 인코딩[['품종_setosa', '품종_versicolor','품종_virginica']]
print(독립.shape, 종속.shape)
```

```python
# 모델 구조 만들기
x = tf.keras.layers.Input(shape=[4])
h = tf.keras.layers.Dense(8, activation='swish')(x)
h = tf.keras.layers.Dense(8, activation='swish')(h)
h = tf.keras.layers.Dense(8, activation='swish')(h)
y = tf.keras.layers.Dense(3, activation='softmax')(h)
model = tf.keras.models.Model(x, y)
model.compile(loss = 'categorical_crossentropy', 
              metrics = 'accuracy')
```

```python
# 모델 학습시키기
model.fit(독립, 종속, epochs=100)
```

```python
# 모델 이용하기
print(model.predict(독립[0:5]))
print(종속[0:5])
```

```python
종속[0:5]
```

```python
# 학습한 가중치
model.get_weights()
```
